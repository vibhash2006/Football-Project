```python
team_id_to_name={}
for id in player_team_ids.values():
    team_id_to_name.setdefault(id,df_26[df_26['team id'] == id]['team'].unique())
```

```python
team_id_to_name = df_26.groupby('team id')['team'].first().to_dict()
```

```python
def apply_2026_overrides(row):
    p_id = row['player id']
    if p_id in player_team_ids:
        target_id = player_team_ids[p_id]
        row['team id'] = target_id
        row['team'] = team_id_to_name.get(target_id, row['team'])
    return row

df_26 = df_26.apply(apply_2026_overrides, axis=1)
```

```python
df_26['rating_points'] = df_26['rating'] * df_26['appearances']
df_25['rating_points'] = df_25['rating'] * df_25['appearances']
df_24['rating_points'] = df_24['rating'] * df_24['appearances']
```

```python
metadata_columns = ['player', 'team', 'player id', 'team id', 'league', 'season', 'position']

percentage_columns = [
    'accurateCrossesPercentage', 'accurateLongBallsPercentage', 'accuratePassesPercentage',
    'aerialDuelsWonPercentage', 'groundDuelsWonPercentage', 'goalConversionPercentage',
    'penaltyConversion', 'setPieceConversion', 'successfulDribblesPercentage',
    'tacklesWonPercentage', 'totalDuelsWonPercentage'
]

special_exclusions = [
    'rating', 'totalRating', 'countRating', 'scoringFrequency'
]
```

```python
percentage_formula_map = {
    'accurateCrossesPercentage': ('accurateCrosses', 'totalCross'),
    'accurateLongBallsPercentage': ('accurateLongBalls', 'totalLongBalls'),
    'accuratePassesPercentage': ('accuratePasses', 'totalPasses'),
    'successfulDribblesPercentage': ('successfulDribbles', 'totalContest'),
    'tacklesWonPercentage': ('tacklesWon', 'tackles'),
    'goalConversionPercentage': ('goals', 'totalShots'),
    
    'penaltyConversion': ('penaltyGoals', 'penaltiesTaken'),
    'setPieceConversion': ('freeKickGoal', 'shotFromSetPiece'),
    
    'aerialDuelsWonPercentage': ('aerialDuelsWon', 'aerialDuelsWon + aerialLost'),
    'groundDuelsWonPercentage': ('groundDuelsWon', 'groundDuelsWon + duelLost'),
    'totalDuelsWonPercentage': ('totalDuelsWon', 'totalDuelsWon + duelLost'),
    
    'scoringFrequency': ('minutesPlayed', 'goals')
}
```

```python
agg_rules = {
    'player': 'last',
    'team': 'last',
    'team id': 'last',
    'league': 'last',
    'season': 'last',
    'position': 'last',
    'appearances': 'sum',
    'rating_points': 'sum' 
}
```

```python
sum_column=df_26.columns.to_list()
for item in special_exclusions + metadata_columns + percentage_columns:    
    sum_column.remove(item)
```

```python
for col in sum_column:
    if col not in agg_rules:
        agg_rules[col] = 'sum'
```

```python
df_26_flat = df_26.groupby('player id').agg(agg_rules).reset_index()
df_25_flat = df_25.groupby('player id').agg(agg_rules).reset_index()
df_24_flat = df_24.groupby('player id').agg(agg_rules).reset_index()
```

```python
for df_flat in [df_26_flat, df_25_flat, df_24_flat]:
    df_flat['rating'] = np.where(
        df_flat['appearances'] > 0,
        df_flat['rating_points'] / df_flat['appearances'],
        0.0
    )
```

```python
df_26_suff = df_26_flat.rename(columns={col: f"{col}".lower()+"_2026" for col in df_26_flat.columns if col != 'player id'})
df_25_suff = df_25_flat.rename(columns={col: f"{col}".lower()+"_2025" for col in df_25_flat.columns if col != 'player id'})
df_24_suff = df_24_flat.rename(columns={col: f"{col}".lower()+"_2024" for col in df_24_flat.columns if col != 'player id'})
```

```python
master_unweighted = pd.merge(df_26_suff, df_25_suff, on='player id', how='left')
master_unweighted = pd.merge(master_unweighted, df_24_suff, on='player id', how='left')
```

```python
historical_metadata_to_drop = [
    'player_2025', 'team_2025', 'team id_2025', 'league_2025', 'season_2025', 'position_2025',
    'player_2024', 'team_2024', 'team id_2024', 'league_2024', 'season_2024', 'position_2024'
]

master_unweighted.drop(columns=historical_metadata_to_drop, inplace=True, errors='ignore')
```

```python
master_unweighted.shape
```

```python
base_metadata = ['player id', 'player_2026', 'team_2026', 'team id_2026', 'league_2026', 'position_2026']
clean_metadata_names = ['player id', 'player', 'team', 'team id', 'league', 'position']

db_average = master_unweighted[base_metadata].copy()
db_average.columns = clean_metadata_names

db_average.head(10)
```

```python
volume_columns = ['appearances', 'rating_points'] + [col.lower() for col in sum_column]

count_2026 = master_unweighted['appearances_2026'].notna().astype(int)
count_2025 = master_unweighted['appearances_2025'].notna().astype(int)
count_2024 = master_unweighted['appearances_2024'].notna().astype(int)
db_average['number_of_seasons'] = count_2026 + count_2025 + count_2024

total_active_years = count_2026 + count_2025 + count_2024

for col in volume_columns:
    total_sum = (
        master_unweighted[f"{col}_2026"].fillna(0) +
        master_unweighted[f"{col}_2025"].fillna(0) +
        master_unweighted[f"{col}_2024"].fillna(0)
    )
    db_average[col] = total_sum / total_active_years
```

```python
db_average['rating'] = np.where(
    db_average['appearances'] > 0, 
    db_average['rating_points'] / db_average['appearances'], 
    0.0
)
db_average.drop(columns=['rating_points'], inplace=True)

for pct_col, (success, total) in percentage_formula_map.items():
    pct_col_lower = pct_col.lower()
    success_lower = success.lower()
    total_lower = total.lower()
    
    if ' + ' in total_lower:
        part1, part2 = total_lower.split(' + ')
        denom = db_average[part1.strip()] + db_average[part2.strip()]
    else:
        denom = db_average[total_lower]
        
    num = db_average[success_lower]
    db_average[pct_col_lower] = np.where(denom > 0, (num / denom) * 100, 0.0)
```

```python
db_average=db_average.drop(columns='scoringfrequency')
```

```python
db_ewma = master_unweighted[base_metadata].copy()
db_ewma.columns = clean_metadata_names
```

```python
mask_2026 = master_unweighted['appearances_2026'].notna().astype(int)
mask_2025 = master_unweighted['appearances_2025'].notna().astype(int)
mask_2024 = master_unweighted['appearances_2024'].notna().astype(int)
db_ewma['number_of_seasons'] = mask_2026 + mask_2025 + mask_2024
```

```python
total_ewma_weight = (1.0 * mask_2026) + (0.5 * mask_2025) + (0.25 * mask_2024)
```

```python
for col in volume_columns:
    weighted_sum = (
        (master_unweighted[f"{col}_2026"].fillna(0) * 1.0) +
        (master_unweighted[f"{col}_2025"].fillna(0) * 0.5) +
        (master_unweighted[f"{col}_2024"].fillna(0) * 0.25)
    )
    
    db_ewma[col] = weighted_sum / total_ewma_weight
```

```python
db_ewma.shape
```

```python
db_ewma['rating'] = np.where(
    db_ewma['appearances'] > 0, 
    db_ewma['rating_points'] / db_ewma['appearances'], 
    0.0
)
db_ewma.drop(columns=['rating_points'], inplace=True)

for pct_col, (success, total) in percentage_formula_map.items():
    pct_col_lower = pct_col.lower()
    success_lower = success.lower()
    total_lower = total.lower()
    
    if ' + ' in total_lower:
        part1, part2 = total_lower.split(' + ')
        denom = db_ewma[part1.strip()] + db_ewma[part2.strip()]
    else:
        denom = db_ewma[total_lower]
        
    num = db_ewma[success_lower]
    db_ewma[pct_col_lower] = np.where(denom > 0, (num / denom) * 100, 0.0)
```

```python
db_ewma=db_ewma.drop(columns='scoringfrequency')
```
