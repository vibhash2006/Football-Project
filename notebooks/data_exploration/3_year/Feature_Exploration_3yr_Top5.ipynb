```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("../../../data/raw/3_year/ThreeYearAvg_Top5_Leagues.csv")
```

```python
df=df.drop(columns=["outfielderblocks",])
```

```python
per90_cols = [
    # Finishing
    "goals",
    "expectedgoals",
    "totalshots",
    "shotsontarget",
    "shotsofftarget",
    "shotsfrominsidethebox",
    "shotsfromoutsidethebox",
    "hitwoodwork",
    "leftfootgoals",
    "rightfootgoals",
    "headedgoals",
    "goalsfrominsidethebox",
    "goalsfromoutsidethebox",
    "freekickgoal",
    "penaltiestaken",
    "penaltygoals",
    "bigchancesmissed",

    # Creativity
    "assists",
    "expectedassists",
    "goalsassistssum",
    "totalpasses",
    "accuratepasses",
    "inaccuratepasses",
    "totaloppositionhalfpasses",
    "accurateoppositionhalfpasses",
    "totalownhalfpasses",
    "accurateownhalfpasses",
    "accuratefinalthirdpasses",
    "keypasses",
    "totalattemptassist",
    "passtoassist",
    "bigchancescreated",
    "totallongballs",
    "accuratelongballs",
    "totalchippedpasses",
    "accuratechippedpasses",
    "totalcross",
    "accuratecrosses",

    # Possession
    "touches",
    "possessionlost",
    "dispossessed",
    "totalcontest",
    "successfuldribbles",
    "offsides",
    "wasfouled",
    "fouls",

    # Defensive
    "tackles",
    "tackleswon",
    "interceptions",
    "ballrecovery",
    "possessionwonattthird",
    "clearances",
    "blockedshots",
    "dribbledpast",
    "errorleadtoshot",
    "errorleadtogoal",

    # Duels
    "totalduelswon",
    "duellost",
    "groundduelswon",
    "aerialduelswon",
    "aeriallost",
    
    # Goalkeeping
    "saves",
    "savescaught",
    "savesparried",
    "savedshotsfrominsidethebox",
    "savedshotsfromoutsidethebox",
    "goalsprevented",
    "runsout",
    "successfulrunsout",
    "goalkicks",
    "punches",
    "highclaims",
    "crossesnotclaimed",  
    

    # Discipline
    "yellowcards",
    "yellowredcards",
    "directredcards",
    "owngoals"
]
```

```python
rate_cols = [
    "goalconversionpercentage",
    "scoringfrequency",

    "accuratepassespercentage",
    "accuratelongballspercentage",
    "successfuldribblespercentage",

    "tackleswonpercentage",

    "totalduelswonpercentage",
    "groundduelswonpercentage",
    "aerialduelswonpercentage",

    "penaltyconversion",

    "rating",
    "totalrating",
    "countrating"
]
```

```python
for col in per90_cols:
    df[f"{col}_per90"] = np.where(
        df["minutesplayed"] > 0,
        (df[col] / df["minutesplayed"]) * 90,
        np.nan
    )
```

```python
df.shape
```

```python
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
df["goals_minus_xg"] = df["goals"] - df["expectedgoals"]
df["assists_minus_xa"] = df["assists"] - df["expectedassists"]
```

```python
df["goals_per_xg"] = np.where(
    df["expectedgoals"] > 0,
    df["goals"] / df["expectedgoals"],
    np.nan
)
```

```python
df["shots_on_target_pct"] = np.where(
    df["totalshots"] > 0,
    df["shotsontarget"] / df["totalshots"],
    np.nan
)
```

```python
df["inside_box_shot_pct"] = np.where(
    df["totalshots"] > 0,
    df["shotsfrominsidethebox"] / df["totalshots"],
    np.nan
)
```

```python
df["assist_conversion"] = np.where(
    df["keypasses"] > 0,
    df["assists"] / df["keypasses"],
    np.nan
)
```

```python
df["xa_per_keypass"] = np.where(
    df["keypasses"] > 0,
    df["expectedassists"] / df["keypasses"],
    np.nan
)
```

```python
df["final_third_pass_pct"] = np.where(
    df["accuratepasses"] > 0,
    df["accuratefinalthirdpasses"] / df["accuratepasses"],
    np.nan
)
```

```python
df["opp_half_pass_pct"] = np.where(
    df["accuratepasses"] > 0,
    df["accurateoppositionhalfpasses"] / df["accuratepasses"],
    np.nan
)
```

```python
df["dribbles_per_touch"] = np.where(
    df["touches"] > 0,
    df["successfuldribbles"] / df["touches"],
    np.nan
)
```

```python
df["dispossessed_per_touch"] = np.where(
    df["touches"] > 0,
    df["dispossessed"] / df["touches"],
    np.nan
)
```

```python
df["possession_lost_per_touch"] = np.where(
    df["touches"] > 0,
    df["possessionlost"] / df["touches"],
    np.nan
)
```

```python
df["defensive_actions"] = (
    df["tackles"] +
    df["interceptions"] +
    df["ballrecovery"]
)
```

```python
df["defensive_actions_per90"] = (
    df["tackles_per90"] +
    df["interceptions_per90"] +
    df["ballrecovery_per90"]
)
```

```python
df['weak_foot_goals_pct'] = np.minimum(
    df['leftfootgoals'], 
    df['rightfootgoals']
)/(df['leftfootgoals']+df['rightfootgoals'])
```
