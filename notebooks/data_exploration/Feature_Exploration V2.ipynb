```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("../../data/raw/v2/sofascore_top5_2526season.csv")
```

```python
df.shape
```

```python
df=df.drop(columns=["outfielderBlocks",])
```

```python
per90_cols = [
    # Finishing
    "goals",
    "expectedGoals",
    "totalShots",
    "shotsOnTarget",
    "shotsOffTarget",
    "shotsFromInsideTheBox",
    "shotsFromOutsideTheBox",
    "hitWoodwork",
    "leftFootGoals",
    "rightFootGoals",
    "headedGoals",
    "goalsFromInsideTheBox",
    "goalsFromOutsideTheBox",
    "freeKickGoal",
    "penaltiesTaken",
    "penaltyGoals",
    "bigChancesMissed",

    # Creativity
    "assists",
    "expectedAssists",
    "goalsAssistsSum",
    "totalPasses",
    "accuratePasses",
    "inaccuratePasses",
    "totalOppositionHalfPasses",
    "accurateOppositionHalfPasses",
    "totalOwnHalfPasses",
    "accurateOwnHalfPasses",
    "accurateFinalThirdPasses",
    "keyPasses",
    "totalAttemptAssist",
    "passToAssist",
    "bigChancesCreated",
    "totalLongBalls",
    "accurateLongBalls",
    "totalChippedPasses",
    "accurateChippedPasses",
    "totalCross",
    "accurateCrosses",

    # Possession
    "touches",
    "possessionLost",
    "dispossessed",
    "totalContest",
    "successfulDribbles",
    "offsides",
    "wasFouled",
    "fouls",

    # Defensive
    "tackles",
    "tacklesWon",
    "interceptions",
    "ballRecovery",
    "possessionWonAttThird",
    "clearances",
    "blockedShots",
    "dribbledPast",
    "errorLeadToShot",
    "errorLeadToGoal",

    # Duels
    "totalDuelsWon",
    "duelLost",
    "groundDuelsWon",
    "aerialDuelsWon",
    "aerialLost",

    # Discipline
    "yellowCards",
    "yellowRedCards",
    "directRedCards",
    "ownGoals"
]
```

```python
rate_cols = [
    "goalConversionPercentage",
    "scoringFrequency",

    "accuratePassesPercentage",
    "accurateLongBallsPercentage",
    "successfulDribblesPercentage",

    "tacklesWonPercentage",

    "totalDuelsWonPercentage",
    "groundDuelsWonPercentage",
    "aerialDuelsWonPercentage",

    "penaltyConversion",

    "rating",
    "totalRating",
    "countRating"
]
```

```python
for col in per90_cols:
    df[f"{col}_per90"] = np.where(
        df["minutesPlayed"] > 0,
        (df[col] / df["minutesPlayed"]) * 90,
        np.nan
    )
```

```python
df.shape
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
df["goals_minus_xg"] = df["goals"] - df["expectedGoals"]
df["assists_minus_xa"] = df["assists"] - df["expectedAssists"]
```

```python
df["goals_per_xg"] = np.where(
    df["expectedGoals"] > 0,
    df["goals"] / df["expectedGoals"],
    np.nan
)
```

```python
df["shots_on_target_pct"] = np.where(
    df["totalShots"] > 0,
    df["shotsOnTarget"] / df["totalShots"],
    np.nan
)
```

```python
df["inside_box_shot_pct"] = np.where(
    df["totalShots"] > 0,
    df["shotsFromInsideTheBox"] / df["totalShots"],
    np.nan
)
```

```python
df["assist_conversion"] = np.where(
    df["keyPasses"] > 0,
    df["assists"] / df["keyPasses"],
    np.nan
)
```

```python
df["xa_per_keypass"] = np.where(
    df["keyPasses"] > 0,
    df["expectedAssists"] / df["keyPasses"],
    np.nan
)
```

```python
df["final_third_pass_pct"] = np.where(
    df["accuratePasses"] > 0,
    df["accurateFinalThirdPasses"] / df["accuratePasses"],
    np.nan
)
```

```python
df["opp_half_pass_pct"] = np.where(
    df["accuratePasses"] > 0,
    df["accurateOppositionHalfPasses"] / df["accuratePasses"],
    np.nan
)
```

```python
df["dribbles_per_touch"] = np.where(
    df["touches"] > 0,
    df["successfulDribbles"] / df["touches"],
    np.nan
)
```

```python
df["dispossessed_per_touch"] = np.where(
    df["touches"] > 0,
    df["dispossessed"] / df["touches"],
    np.nan
)
```

```python
df["possession_lost_per_touch"] = np.where(
    df["touches"] > 0,
    df["possessionLost"] / df["touches"],
    np.nan
)
```

```python
df["defensive_actions"] = (
    df["tackles"] +
    df["interceptions"] +
    df["ballRecovery"]
)
```

```python
df["defensive_actions_per90"] = (
    df["tackles_per90"] +
    df["interceptions_per90"] +
    df["ballRecovery_per90"]
)
```

```python
df['weak_foot_goals_pct'] = np.minimum(
    df['leftFootGoals'], 
    df['rightFootGoals']
)/(df['leftFootGoals']+df['rightFootGoals'])
```
