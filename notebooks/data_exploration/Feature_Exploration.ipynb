# Feature Creation #

```python
df["goals_minus_xg"] = (df["Performance_Gls"] - df["xg"])
```

```python
df["shot_conversion"] = (df["Performance_Gls"] / df["Standard_Sh"])
```

```python
df["sot_ratio"] = (df["Standard_SoT"] / df["Standard_Sh"])
```

```python
df["xg_per_shot"] = (df["xg"] / df["Standard_Sh"])
```

```python
df["xa_per_key_pass"] = (df["xa"] / df["key_passes"])
```

```python
df["assist_minus_xa"] = (df["Performance_Ast"] - df["xa"])
```

```python
df["buildup_share"] = (df["xg_buildup"] / df["xg_chain"])
```

```python
df["finishing_share"] = (df["xg"] / df["xg_chain"])
```

```python
df["xg_per90"] = df["xg"] / df["90s"]
df["xa_per90"] = df["xa"] / df["90s"]
df["shots_per90"] = df["shots"] / df["90s"]
df["key_passes_per90"] = (df["key_passes"] / df["90s"])
df["xg_chain_per90"] = (df["xg_chain"] / df["90s"])
df["xg_buildup_per90"] = (df["xg_buildup"] / df["90s"])
```
