```python
df=df.drop(columns=['loser_team','winner_team','matchday','match_result','home_goals','away_goals','match_id','date'])
```

```python
df.shape
```

```python
count=0
for idx in df.index:
    if count%2==0:
        home = df.at[idx, 'home_team']
        away = df.at[idx, 'away_team']
        
        df.at[idx, 'home_team'] = away
        df.at[idx, 'away_team'] = home
        
        df.at[idx, 'goal_difference'] = -df.at[idx, 'goal_difference']
        
    count+=1
```
