```python
from ScraperFC.fbref import FBref

fb = FBref()

data = fb.scrape_stats(
    year="2024-2025",
    league="England Premier League",
    stat_category="passing"
)

print(data.keys())
```

```python
player_df = data["player"]

print(player_df.shape)
print(player_df.columns.tolist())
player_df.head()
```

```python
print(player_df.columns)
```

```python
player_df.columns = [
    "_".join([str(x) for x in col if "Unnamed" not in str(x)])
    for col in player_df.columns
]

print(player_df.columns.tolist())
```

```python
player_df.isna().sum()
```

```python
for cat in ["standard", "shooting", "passing","possesion"]:
    data = fb.scrape_stats(
        year="2024-2025",
        league="England Premier League",
        stat_category=cat
    )

    df = data["player"]

    print(cat)
    print(df.shape)
    print(df.isna().sum().sum())
```
