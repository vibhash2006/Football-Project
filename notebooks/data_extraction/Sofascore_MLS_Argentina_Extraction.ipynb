```python
import time
import os
import pandas as pd
import ScraperFC as sfc
```

```python
output_dir = "../../data/raw/major_leagues v1"
```

```python
ss=sfc.Sofascore()
```

```python
ss.get_valid_seasons("Argentina Liga Profesional")
```

```python
Seasons=["2026","2025","2024"]
Leagues = ["USA MLS","Argentina Liga Profesional"]
Positions = ["Goalkeepers", "Defenders", "Midfielders", "Forwards"]
```

```python
for season in Seasons:
    master_df = []

    for league in Leagues:
        for pos in Positions:

            print(f"Scraping {league} {season} {pos}")

            try:
                df = ss.scrape_player_league_stats(
                    year=season,
                    league=league,
                    accumulation="total",
                    selected_positions=[pos]
                )

                df["league"] = league
                df["season"] = season
                df["position"] = pos

                master_df.append(df)

                print(f"{len(df)} players, {len(df.columns)} columns")

            except Exception as e:
                print(f"fail: {e}")

            time.sleep(5)

    if master_df:

        final_season_df = pd.concat(
            master_df,
            ignore_index=True
        )

        file_path = (
            f"{output_dir}/"
            f"sofascore_MLS_Argentina_{season[2:]}season.csv"
        )

        final_season_df.to_csv(
            file_path,
            index=False
        )

        print(
            f"Saved: {len(final_season_df)} total rows for {season} season"
        )

    else:

        print(
            f"No data was scraped for the {season} season."
        )
```
