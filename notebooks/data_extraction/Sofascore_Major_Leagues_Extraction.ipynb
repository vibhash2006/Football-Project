```python
import time
import os
import pandas as pd
import ScraperFC as sfc
```

```python
output_dir = "../../data/raw/major_leagues v1"
```

```python
ss=sfc.Sofascore()
```

```python
Seasons=["25/26","24/25","23/24"]
Leagues = [
    "England Premier League","Spain La Liga","Germany Bundesliga",
    "Italy Serie A","France Ligue 1","England EFL Championship",
    "Spain La Liga 2","Germany 2.Bundesliga","Italy Serie B","France Ligue 2",
    "Netherlands Eredivisie","Portugal Primeira Liga","Turkiye Super Lig",
    "Saudi Arabia Pro League",
]
Positions = ["Goalkeepers", "Defenders", "Midfielders", "Forwards"]
```

```python
for season in Seasons:
    master_df = []

    for league in Leagues:
        for pos in Positions:

            print(f"Scraping {league} {season} {pos}")

            try:
                df = ss.scrape_player_league_stats(
                    year=season,
                    league=league,
                    accumulation="total",
                    selected_positions=[pos]
                )

                df["league"] = league
                df["season"] = season
                df["position"] = pos

                master_df.append(df)

                print(f"{len(df)} players, {len(df.columns)} columns")

            except Exception as e:
                print(f"fail: {e}")

            time.sleep(5)

    if master_df:

        final_season_df = pd.concat(
            master_df,
            ignore_index=True
        )

        file_path = (
            f"{output_dir}/"
            f"sofascore_major_leagues_{season[0:2] + season[3:]}season.csv"
        )

        final_season_df.to_csv(
            file_path,
            index=False
        )

        print(
            f"Saved: {len(final_season_df)} total rows for {season} season"
        )

    else:

        print(
            f"No data was scraped for the {season} season."
        )
```

# The 16 leagues were decided based on availability of data and number of players playing in the world cup per league. The Second Divison Leagues were choosen for future predictions of the top 5 leagues #
