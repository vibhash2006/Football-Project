```python
import soccerdata as sd
import pandas as pd

fbref = sd.FBref(leagues="ENG-Premier League",
    seasons=2025)
```

```python
df = fbref.read_player_season_stats(stat_type="standard")
```

```python
print(df.columns)
print(df.index.names)
```

```python
# move index into columns
df = df.reset_index()

# flatten multiindex columns
df.columns = [
    "_".join(col).strip("_")
    if isinstance(col, tuple)
    else col
    for col in df.columns
]
# Inspect available columns first
print(df.columns)
```

```python
strikers = df[
    df["pos"].str.contains("FW", na=False)
]

strikers = strikers[
    strikers["Playing Time_Min"] >= 500
]

strikers = strikers.sort_values(
    by="Performance_Gls",
    ascending=False
)

print(strikers)
```
