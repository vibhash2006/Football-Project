```python
team_id_to_name={}
for id in player_team_ids.values():
    team_id_to_name.setdefault(id,df_26[df_26['team id'] == id]['team'].unique())
```

```python
team_id_to_name = df_26.groupby('team id')['team'].first().to_dict()
```

```python
def apply_2026_overrides(row):
    p_id = row['player id']
    if p_id in player_team_ids:
        target_id = player_team_ids[p_id]
        row['team id'] = target_id
        row['team'] = team_id_to_name.get(target_id, row['team'])
    return row

df_26 = df_26.apply(apply_2026_overrides, axis=1)
```

```python
df_26['rating_points'] = df_26['rating'] * df_26['appearances']
df_25['rating_points'] = df_25['rating'] * df_25['appearances']
df_24['rating_points'] = df_24['rating'] * df_24['appearances']
```

```python
metadata_columns = ['player', 'team', 'player id', 'team id', 'league', 'season', 'position']

percentage_columns = [
    'accurateCrossesPercentage', 'accurateLongBallsPercentage', 'accuratePassesPercentage',
    'aerialDuelsWonPercentage', 'groundDuelsWonPercentage', 'goalConversionPercentage',
    'penaltyConversion', 'setPieceConversion', 'successfulDribblesPercentage',
    'tacklesWonPercentage', 'totalDuelsWonPercentage'
]

special_exclusions = [
    'rating', 'totalRating', 'countRating', 'scoringFrequency'
]
```

```python
percentage_formula_map = {
    'accurateCrossesPercentage': ('accurateCrosses', 'totalCross'),
    'accurateLongBallsPercentage': ('accurateLongBalls', 'totalLongBalls'),
    'accuratePassesPercentage': ('accuratePasses', 'totalPasses'),
    'successfulDribblesPercentage': ('successfulDribbles', 'totalContest'),
    'tacklesWonPercentage': ('tacklesWon', 'tackles'),
    'goalConversionPercentage': ('goals', 'totalShots'),
    
    'penaltyConversion': ('penaltyGoals', 'penaltiesTaken'),
    'setPieceConversion': ('freeKickGoal', 'shotFromSetPiece'),
    
    'aerialDuelsWonPercentage': ('aerialDuelsWon', 'aerialDuelsWon + aerialLost'),
    'groundDuelsWonPercentage': ('groundDuelsWon', 'groundDuelsWon + duelLost'),
    'totalDuelsWonPercentage': ('totalDuelsWon', 'totalDuelsWon + duelLost'),
    
    'scoringFrequency': ('minutesPlayed', 'goals')
}
```

```python
agg_rules = {
    'player': 'last',
    'team': 'last',
    'team id': 'last',
    'league': 'last',
    'season': 'last',
    'position': 'last',
    'appearances': 'sum',
    'rating_points': 'sum' 
}
```

```python
sum_column=df_26.columns.to_list()
for item in special_exclusions + metadata_columns + percentage_columns:    
    sum_column.remove(item)
```

```python
for col in sum_column:
    if col not in agg_rules:
        agg_rules[col] = 'sum'
```

```python
df_26_flat = df_26.groupby('player id').agg(agg_rules).reset_index()
df_25_flat = df_25.groupby('player id').agg(agg_rules).reset_index()
df_24_flat = df_24.groupby('player id').agg(agg_rules).reset_index()
```

```python
for df_flat in [df_26_flat, df_25_flat, df_24_flat]:
    df_flat['rating'] = np.where(
        df_flat['appearances'] > 0,
        df_flat['rating_points'] / df_flat['appearances'],
        0.0
    )
```
